```r
library(tidyverse)
```

```r
rand_vals <- rnorm(10)
```

```r
hist(
    rand_vals,
    col = 'green',
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

```r
rand_vals_500 <- rnorm(500)
```

```r
rand_vals_50000 <- rnorm(50000)
```

```r
hist(
    rand_vals_500,
    col = 'green',
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

```r
hist(
    rand_vals_50000,
    col = 'green',
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

For the histogram with 10 random numbers, the width of the class was 1/2. For the histogram with 500 random values, the width of the class was 1/2. For the histogram with 50000 random values, the width of the class was also 1/2.

```r
hist(
    rand_vals,
    col = 'green',
    breaks = 50,
    freq = FALSE,
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

```r
hist(
    rand_vals_500,
    col = 'green',
    breaks = 50,
    freq = FALSE,
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

```r
hist(
    rand_vals_50000,
    col = 'green',
    breaks = 50,
    freq = FALSE,
    xlab = 'Values',
    main = 'Random Normal Data'
    )
```

As I increase the size of the data as well as the number of bins, the data becomes very uniform and is a standard bell curve around the median of 0. The shape being created looks very much like a unimodal distribution of data. I think as the size of the data goes to infinity and the class width goes to 0, the shape and density will stay mostly the same but the number of bins will change.

```r
treated <- filter(Puromycin, state == "treated")
```

```r
treated
```

```r
hist(
    treated$rate,
    xlab = 'Treated rate',
    main = 'Treated Cells'
    )
```

```r
untreated <- filter(Puromycin, state == "untreated")
```

```r
hist(
    untreated$rate,
    xlab = 'Untreated rate',
    main = 'Untreated Cells'
    )
```

Based on the histograms of the two data sets of treated cells and untreated cells, my conjecture on the affect of Puromycin on reaction rate was that Puromycin accelerated the reaction rate of treated cells since treated cells have a higher average and maximum reaction rate than the cells that were untreated.

```r
hist(
    treated$rate,
    breaks = seq(0,250,62.5),
    xlab = 'Treated rate',
    main = 'Wide classes for treated cells'
    )
```

```r
hist(
    treated$rate,
    breaks = seq(0,250,10),
    xlab = 'Treated rate',
    main = 'Narrow classes for treated cells'
    )
```

Of the three histograms for the treated cells, the one I feel gives the most useful summary of the data is the one from part two (the original/default one) because it gives the general shape of the data and has a good number of bins. It shows the mode of the data while also not showing the exact rates like the narrow class histogram does, which may make it harder and take more time to understand the data.
