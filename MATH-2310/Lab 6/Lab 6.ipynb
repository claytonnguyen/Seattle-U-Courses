```r
movies <- read.csv("Movies.csv")
movies
```

```r
mov <- movies %>% 
select(Title, Year, Distributor, Type, Budget, IMDB.Rating, Opening.Theaters, Opening.Revenue) %>%
    mutate(
        profit = Opening.Revenue - Budget,
        disney = Distributor == "Walt Disney"
    )
```

```r
mov
```

```r
movie <- mov %>%
  mutate(
    profit = cut(profit,
      breaks = c(-Inf,0,Inf),
      labels = c("no profit", "profit")
) )
```

```r
opening_profit <- table(movie$profit, movie$disney)%>% prop.table()
opening_profit
```

```r
profit_given_disney <- opening_profit[2,2]/(opening_profit[1,2]+opening_profit[2,2])
```

```r
profit_given_disney
```

The conditional probability of making a profit during opening given that it is a Disney film is roughly 6.89%

```r
chisq.test(movie$profit, movie$disney)
```

My interpretation of the results of the previous part is that since the p-value is not small (less than 5%), I can't say for certain whether the two variables are independent or not. So I know neither if the variables are independent or not. However the p value is certainly closer to 5% so I would bet that the two variables are independent of each other.
