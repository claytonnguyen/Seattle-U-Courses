```r
ggplot(stocks_data,aes(sample=open))+
    stat_qq(color="blue", distribution = qexp)+
    stat_qq_line(color="red", distribution = qexp)
```

```r
cor_open_volume <- cor(stocks$open, stocks$volume)
cor_open_volume
```

```r
stocks <- stocks %>%
  mutate(
    fracgrp = cut(open,
      breaks = c(0, 0.000223, 0.001991, 0.002802, 1),
      labels = c("Q1", "Q2", "Q3", "Q4")
) )
```

```r
ggplot(stocks,aes(x = open,y = volume))+
    geom_point(aes(color=fracgrp))+
    geom_smooth(method = 'lm',aes(color=fracgrp), se = FALSE)
```

```r
summary(lm(volume~open+fracgrp, stocks))
```

I dont think the model is a good fit since the rsquared value is not that close to 1, I think this has to do with the exponential jump in dogecoin's price in the past month, a very small sample skewing the rest of the data population.
